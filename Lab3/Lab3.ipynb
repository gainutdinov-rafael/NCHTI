##Выполнение задания:

```python
from concurrent import futures
```

###Объявление функций:

```python
def function1():
  for i in range(5):
    print(f"Function 1 : {i}")
  print("Function 1 done!")

def function2():
  for i in range(5):
    print(f"Function 2 : {i}")
  print("Function 2 done!")
```

```python
executer = futures.ThreadPoolExecutor()
future1 = executer.submit(function1)
future2 = executer.submit(function2)
# Извлечение результатов
result1 = future1.result()
result2 = future2.result()
# Вывод результатов
print(f"{result1}\n{result2}")
executer.shutdown()
```
